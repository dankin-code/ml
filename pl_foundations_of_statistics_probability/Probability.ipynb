### Probability
#### The extent to which an event is likely to occur, meausured by the ratio of the favorable cases to the whole number of cases possible.
#### Gaussian distribution - properties in the real world can be represented by a normal distribution. Or the bell curve.

```python
import random
import numpy as np
import scipy.stats
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")
```

```python
# define a functions that simulates coin flip
def flip(n):
    result = []
    for i in range( 1, n+1):
        result.append(random.choice(['Heads','Tails']))
    return result
```

```python
result = flip(10)
result
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x=result)

plt.xlabel('Coin Flip')
plt.title('Counts')

plt.show()
```

```python
result = flip(100)
plt.figure(figsize=(12,8))
sns.countplot(x=result)

plt.xlabel('Coin Flip')
plt.title('Counts')

plt.show()
```

```python
result = flip(100000)
plt.figure(figsize=(12,8))
sns.countplot(x=result)

plt.xlabel('Coin Flip')
plt.title('Counts')

plt.show()
```
