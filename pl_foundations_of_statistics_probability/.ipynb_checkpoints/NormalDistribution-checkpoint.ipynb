```python
import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.api as sm
from scipy.stats import norm
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style='darkgrid')
```

```python
normally_distributed_data = np.random.normal(loc=0, scale=1, size=100000)
normally_distributed_data[:10]
```

```python
sample_mean = np.mean(normally_distributed_data)
sample_mean
```

```python
sample_std = np.std(normally_distributed_data)
sample_std
```

```python
sample_med = np.median(normally_distributed_data)
sample_med
```

```python
plt.figure(figsize=(12,8))

sns.histplot(x=normally_distributed_data)

plt.title('Normal Distribution', fontsize='14')

plt.axvline(x=sample_mean, color='yellow', linestyle='-')
plt.axvline(x=sample_med, color='red', linestyle='-.')

plt.xlabel('Values', fontsize='12')
plt.ylabel('Probability', fontsize='12')

plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.kdeplot(x=normally_distributed_data)

plt.title('Normal Distribution', fontsize='14')

plt.axvline(x=sample_mean, color='yellow', linestyle='-')
plt.axvline(x=sample_med, color='red', linestyle='-.')

plt.xlabel('Values', fontsize='12')
plt.ylabel('Probability', fontsize='12')

plt.show()
```

```python
#loc0 = mean 0
#scale3 = standard deviation 3
normally_distributed_data_scale3 = np.random.normal(loc=0, scale=3, size=100000)

plt.figure(figsize=(12,8))

sns.kdeplot(x=normally_distributed_data)
sns.kdeplot(x=normally_distributed_data_scale3)

plt.title('Normal Distribution', fontsize='14')

plt.axvline(x=sample_mean, color='yellow', linestyle='-')
plt.axvline(x=sample_med, color='red', linestyle='-.')

plt.xlabel('X', fontsize='12')
plt.ylabel('Probability', fontsize='12')

plt.show()
```

```python
#loc2 = mean 2
#scale1 = standard deviation 1
normally_distributed_data_loc2_scale1 = np.random.normal(loc=2, scale=1, size=100000)

plt.figure(figsize=(12,8))

sns.kdeplot(x=normally_distributed_data)
sns.kdeplot(x=normally_distributed_data_loc2_scale1)

plt.title('Normal Distribution', fontsize='14')
plt.xlabel('X', fontsize='12')
plt.ylabel('Probability', fontsize='12')

plt.axvline(x=sample_mean, color='yellow', linestyle='-')
plt.axvline(x=sample_med, color='red', linestyle='-.')

plt.show()
```
